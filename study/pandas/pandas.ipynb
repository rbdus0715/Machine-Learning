## **2. dataframe <-> ndarray, list, dict 변환**

```python
import numpy as np
```

배열, 딕셔너리 --> 데이터프레임

```python
# 2행 3열의 list, ndarray를 dataframe으로 변환하기
col_name = ['col1', 'col2', 'col2']

list = [
    [1, 2, 3],
    [11, 12, 13]
]

array = np.array(list)

df_list = pd.DataFrame(list, columns=col_name)
df_ndarray = pd.DataFrame(array, columns=col_name)

print(df_list, '\n')
print(df_ndarray)
```

```python
# 딕셔너리를 데이터프레임으로 변환하기
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df = pd.DataFrame(dict)
print(df)
```

데이터프레임 --> 배열, 딕셔너리

```python
# ndarray로 변환하기
arr = df.values
print(arr, '\n')

# 리스트로 변환하기
list_ = df.values.tolist()
print(list_, '\n')

# 딕셔너리로 변환하기
dict_ = df.to_dict('list')
print(dict_)
```

## **6. 정렬, Aggregation, GroupBy**
- 정렬 : sort_values(by=[ 칼럼 이름,... ],  ascending=True)
- Aggregation : DataFrame이 특정 칼럼들만을 추출해 적용한다.
    - aggregation 함수들 : min( ), max( ), sum( ), count( )
    - ex) t_df[ [ 'Age' ,  'Fare' ] ].mean( )
- Groupby : 원하는 칼럼을 기준으로 그룹화하기
    - **다양한 사용법들**
    - dataFrame.groupby('칼럼')[ 다른 칼럼, ... ].count()
    - dataFrame.groupby('칼럼')[ 다른 칼럼 ].agg( [ max, min ] )
    - dataFrame.groupby('칼럼').agg( { '칼럼1' : 'min',  '칼럼2' : 'max' } )

```python
# (1) 정렬

titanic_sorted = t_df.sort_values(by=['Name'])
titanic_sorted.head(3)
```

```python
titanic_sorted = t_df.sort_values(by=['Pclass', 'Name'], ascending=False)
titanic_sorted.head(3)
```

```python
# (2) aggregation

t_df[['Age', 'Fare']].mean()
```

```python
# (3) groupby()
# : 원하는 칼럼을 기준으로 나눌 수 있다.
t_groupby = t_df.groupby(by='Pclass')
t_groupby.count()
```

```python
# groupby 이후에 인덱스 선택을 통해 원하는 칼럼만 가져와 원하는 변수 간의 관계를 파악할 수 있다.
t_groupby = t_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
t_groupby
```

```python
t_df.groupby('Pclass')['Age'].agg([max, min])
```

```python
agg_format = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
t_df.groupby('Pclass').agg(agg_format)
```

## **8. apply lambda 식으로 데이터 가공**

- 람다 : 함수 선언과 함수 내의 처리를 한 줄의 식으로 간단히 변환한 식
    - 형태 : lambda (입력 인자) : (입력 인자를 기반으로 한 계산식)
    - 데이터프레임의 원하는 칼럼에 대하여 **DataFrame[ 'col1' ].apply(람다식)** 을 통해 적용할 수 있다.

```python
lambda_square = lambda x : x ** 2
print(lambda_square(3))
```

```python
t_df['Name_len'] = t_df['Name'].apply(lambda x : len(x))
t_df[['Name', 'Name_len']].head(3)
```
