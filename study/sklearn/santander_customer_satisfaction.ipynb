```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head()
```

```python
df['var3'].replace(-999999, 2, inplace=True)
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=0)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=500, learning_rate=0.1, random_state=156, max_depth=3)
xgb.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_train, y_train), (X_test, y_test)])
pred = xgb.predict(X_test)
pred_proba = xgb.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import precision_recall_curve

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred_proba)

    print('오차 행렬\n', confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, AUC: {4:.4f}\n'.format(accuracy, precision, recall, f1, roc_auc))

get_clf_eval(y_test, pred, pred_proba)
```

```python
from sklearn.model_selection import GridSearchCV

xgb = XGBClassifier(n_estimators=100)

params = {
    'max_depth':[5,7],
    'min_child_weight':[1,3],
    'colsample_bytree':[0.5, 0.75]
}

gridcv = GridSearchCV(xgb, param_grid=params, cv=3)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc', eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
pred = gridcv.predict(X_test)
pred_proba = gridcv.predict_proba(X_test)[:, 1]

get_clf_eval(y_test, pred, pred_proba)
```

```python
gridcv.best_params_
```

같은 하이퍼 파라미터 조건에서 learning_rate 0.02로 감소하고 n_estimators=1000으로 증가, reg_alpha=0.03 추가

```python
xgb = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, max_depth=5, min_child_weight=3, colsample_bytree=0.5, reg_alpha=0.03)
xgb.fit(X_train, y_train, early_stopping_rounds=200, eval_metric='auc', eval_set=[(X_train, y_train),(X_test, y_test)])

pred = gridcv.predict(X_test)
pred_proba = gridcv.predict_proba(X_test)[:, 1]

get_clf_eval(y_test, pred, pred_proba)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 12))
plot_importance(xgb, ax=ax, max_num_features=20, height=0.4)
```
