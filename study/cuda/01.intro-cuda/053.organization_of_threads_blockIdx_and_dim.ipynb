### **(1) blockIdx (type: dim3)**
- 그리드(grid)안의 스레드 블럭에게 부여되는 인덱스로, 같은 블럭 안의 스레드들의 blockIdx는 모두 같다.

![image](https://acenet-arc.github.io/ACENET_Summer_School_GPGPU/fig/cuda_indexing.png)

### **(2) blockDim (type: dim3)**
- 스레드블럭의 차원 값을 담고있다.
- 그리드 안의 모든 스레드 블럭은 같은 차원값을 갖는다. => 그리드 안의 모든 스레드에 대해서 모두 같다.

### **(3) GridDim (type: dim3)**
- grid에서 각각의 차원에 대해 몇 개의 스레드 블럭이 있는지

### **(4) 실습 : 256 threads**

```python
%%cu
#include "cuda_runtime.h"
#include "device_launch_parameters.h"
#include <stdio.h>

// blockIdx, blockDim, gridDim 출력
__global__ void print_details() {
    printf("blockIdx.x : %d, blockIdx.y : %d, blockIdx.z : %d, blockDim.x : %d, blockDim.y : %d, gridDim.x : %d, gridDim.y : %d\n",
           blockIdx.x, blockIdx.y, blockIdx.z, blockDim.x, blockDim.y, gridDim.x, gridDim.y);
}

int main() {
    int nx, ny;
    nx = 16;
    ny = 16;

    dim3 block(8, 8);
    dim3 grid(nx/block.x, ny/block.y);

    print_details<<<grid, block>>>();
    cudaDeviceSynchronize();
    ;
    cudaDeviceReset();
    return 0;
}
```
